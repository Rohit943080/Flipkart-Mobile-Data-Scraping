```python
# code here
from bs4 import BeautifulSoup 
import re
import time
import requests
import pandas as pd

headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

url = f'https://www.flipkart.com/search?q=smartphones&page={1}'
response = requests.get(url, headers = headers)

soup = BeautifulSoup(response.text)



final_df = pd.DataFrame()

# This might take some time because we are scraping the data from 389 pages .

for i in range(1, 389):
  url = f'https://www.flipkart.com/search?q=smartphones&page={i}'
  time.sleep(3)
  response = requests.get(url, headers = headers)

  soup = BeautifulSoup(response.text)
  image_url = []
  image_name = []
  avg_rating = []
  total_rating = []
  discount_price = []
  actual_price = []
  reviews = []

  containers = soup.find_all('div', class_ = 'tUxRFH')
  for container in containers:
    image_url.append((container.find('img').get('src')))
    image_name.append((container.find('img').get('alt')))

    # appending average rating
    try:
      avg_rating.append(container.find('div', class_= 'XQDdHH').text)
    except:
      avg_rating.append('')

    # appending total rating and reviews
    try:
      rating , review = container.find('span', class_ = 'Wphh3N').text.strip().split('&') 
      total_rating.append(re.sub(r'\D', '', rating))
      reviews.append(re.sub(r'\D', '', review))
    except:
      total_rating.append('')
      reviews.append('')

      
    # appending dicount price
    try:
        discount_price.append(container.find('div', class_ = 'Nx9bqj _4b5DiR').text)
    except:
        discount_price.append('')

    # appening actual price

    try:
      actual_price.append(container.find('div', class_ = 'yRaY8j ZYYwLA').text)
    
    except:
      actual_price.append('')


  df = pd.DataFrame({
      'image_url' : image_url,
      'name' : image_name,
      'average_rating': avg_rating,
      'total_rating' : total_rating,
      'total_reviews' : reviews,
      'discount_price' : discount_price,
      'actual_price' : actual_price
  })
    

  # appending pages
  final_df = pd.concat((final_df, df))
```
